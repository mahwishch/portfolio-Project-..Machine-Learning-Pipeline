```python
import gradio as gr
import requests

def predict_medical_charges(age, bmi, children, sex, smoker, region):
    url = "http://0.0.0.0:8000/predict/"
    data = {
        "age": age,
        "bmi": bmi,
        "children": children,
        "sex": sex,
        "smoker": smoker,
        "region": region
    }
    response = requests.post(url, json=data)
    prediction = response.json()
    return prediction["prediction"]

iface = gr.Interface(
    fn=predict_medical_charges,
    inputs=[
        gr.Slider(minimum=0, maximum=120, step=1, label="Age"),
        gr.Slider(minimum=10, maximum=50, step=0.1, label="BMI"),
        gr.Slider(minimum=0, maximum=10, step=1, label="Children"),
        gr.Dropdown(choices=["male", "female"], label="Sex"),
        gr.Dropdown(choices=["yes", "no"], label="Smoker"),
        gr.Dropdown(choices=["northwest", "northeast", "southwest", "southeast"], label="Region")
    ],
    outputs="text",
    title="Medical Charges Predictor",
    description="Enter personal attributes to estimate medical charges."
)

if __name__ == "__main__":
    iface.launch()
```
